```python
sale['Sale_Date'] = pd.to_datetime(sale['Sale_Date'])
sale = sale.sort_values('Sale_Date')
```

```python

sale_daily = sale.groupby('Sale_Date')['Sales_Amount'].sum()
sale_daily = sale_daily.asfreq('D')
sale_daily = sale_daily.ffill()
```

```python
plt.figure(figsize=(10,5))
plt.plot(sale_daily)
plt.title("Daily Sales Trend")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()
```

```python
split = int(len(sale_daily) * 0.8)

train = sale_daily[:split]
test = sale_daily[split:]
```

```python
sale['day_of_week'] = sale['Sale_Date'].dt.dayofweek
sale['month'] = sale['Sale_Date'].dt.month
sale['is_weekend'] = sale['day_of_week'].isin([5, 6])
```

```python
day_sales = sale.groupby('day_of_week')['Sales_Amount'].mean()

plt.figure(figsize=(8,4))
day_sales.plot(kind='bar')
plt.title("Average Sales by Day of Week")
plt.xlabel("Day (0=Mon, 6=Sun)")
plt.ylabel("Average Sales")
plt.show()
```

```python
month_sales = sale.groupby('month')['Sales_Amount'].mean()

plt.figure(figsize=(8,4))
month_sales.plot(kind='bar')
plt.title("Average Sales by Month")
plt.xlabel("Month")
plt.ylabel("Average Sales")
plt.show()

```

```python
weekend_sales = sale.groupby('is_weekend')['Sales_Amount'].mean()

plt.figure(figsize=(6,4))
weekend_sales.plot(kind='bar')
plt.title("Weekend vs Weekday Sales")
plt.xlabel("Is Weekend")
plt.ylabel("Average Sales")
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error
```

```python
arima_model = ARIMA(train, order=(1,1,1))
arima_result = arima_model.fit()

arima_pred = arima_result.forecast(steps=len(test))

arima_rmse = np.sqrt(mean_squared_error(test, arima_pred))
print("\nARIMA RMSE:", round(arima_rmse, 2))

```

```python
sarima_model = SARIMAX(train, order=(1,1,1), seasonal_order=(1,1,1,7))
sarima_result = sarima_model.fit()

sarima_pred = sarima_result.forecast(steps=len(test))

sarima_rmse = np.sqrt(mean_squared_error(test, sarima_pred))
print("SARIMA RMSE:", round(sarima_rmse, 2))
```

```python
if arima_rmse < sarima_rmse:
    print("ARIMA performed better")
else:
    print("SARIMA performed better")
```

```python
plt.figure(figsize=(10,5))

plt.plot(test.values, label='Actual')
plt.plot(arima_pred, label='ARIMA Prediction')
plt.plot(sarima_pred, label='SARIMA Prediction')

plt.title("Model Comparison: Actual vs Predictions")
plt.legend()
plt.show()
```

```python
future_steps = 30

future_forecast = arima_result.forecast(steps=future_steps)

plt.figure(figsize=(10,5))
plt.plot(sale_daily[-50:], label='Recent Sales')
plt.plot(range(len(sale_daily), len(sale_daily)+future_steps),
         future_forecast, label='Future Forecast')

plt.title("Next 30 Days Sales Forecast")
```
